```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\vshuk\OneDrive\Desktop\inHOUSEpractical\smart wearables\datasets\cStick\cStick.csv')
df=data
```

```python
df.rename({"Distance":"DISTANCE","Pressure":"PRESSURE","Sugar level":"SUGAR_LEVEL","SpO2":"SPO2","Accelerometer":"ACCELEROMETER","result":"RESULT"},axis=1, inplace=True)
```

```python
X=df.iloc[:,0:5]
y=df.iloc[:,6]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
def linregg(df):
    model=LinearRegression()
    print("Linear Regression model parameters")
    print(model.fit(X_train,y_train))
    print("model accuracy -> ",model.score(X_test,y_test))   
```

```python
def logregg(df):
    model=LogisticRegression(solver='liblinear', C=10.0, random_state=0)
    print("Logistic Regression model parameters")
    print(model.fit(X_train,y_train))
    print("model accuracy -> ",model.score(X_test,y_test))   
```

```python
def Rforest(df):
    model=RandomForestClassifier()
    print("Random Forest Classifier model parameters")
    print(model.fit(X_train,y_train))
    print("model accuracy -> ",model.score(X_test,y_test))   
```

```python
def Dtree(df):
    model=DecisionTreeClassifier()
    print("Decision Tree Classifier model parameters")
    print(model.fit(X_train,y_train))
    print("model accuracy -> ",model.score(X_test,y_test))   
```

```python
w=widgets.Dropdown(options=['Linear Regression','Logistic Regression', 'Decision Tree Classifier','Random Forest Classifier']
                   ,value='Linear Regression')
@interact(model=w)
def printmodel(model):
    if model=='Linear Regression':
        linregg(df)
    elif model=='Logistic Regression':
        logregg(df)
    elif model=='Random Forest Classifier':
        Rforest(df)
    elif model=="Decision Tree Classifier":
        Dtree(df)
```

```python
val=widgets.IntSlider()
@interact(rows=val)
def datasetdisplay(rows):
    print("Dataset used ------")
    print(df[:rows])
```

```python
def norplot(df,col1,col2):
    colmn1=[]
    colmn2=[]
    f = plt.figure()
    f.set_figwidth(12)
    f.set_figheight(8)
    flag=1    
    for x in df[col1]:
        colmn1.append(x)
    for x in df[col2]:
        colmn2.append(x)
    print(len(colmn1))
    print(len(colmn2))
    plt.bar(colmn1,colmn2)
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.title("Barplot of "+col1+" and "+col2)
    plt.grid(axis='y')
    plt.show()
```
